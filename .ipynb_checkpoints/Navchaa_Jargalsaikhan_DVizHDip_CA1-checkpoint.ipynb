```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
from scipy import stats
sns.set()
import warnings
warnings.filterwarnings("ignore")
```

```python
df= pd.read_csv("vgsales.csv")
```

```python
df.head(10)
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
top_5=pd.crosstab(index=df["Global_Sales"], columns="count")
```

```python
sorted_df = df.sort_values(by='Global_Sales', ascending=False)
```

```python
top_5_sales = sorted_df.head(5)
```

```python
print("Top 5 global sales:")
print(top_5_sales[['Genre', 'Global_Sales']])
```

```python
sns.set(style="whitegrid")  # Set the style
plt.figure(figsize=(10, 6))  # Set the figure size

sns.barplot(x='Genre', y='Global_Sales', data=df, ci=None)  # Customize as needed
plt.xlabel('Genre')
plt.ylabel('Global Sales')
plt.title('Average Global Sales by Genre')

plt.show()  # Display the plot
```

```python
top_5_sales_by_genre = df.sort_values(by='Global_Sales', ascending=False).head(7)
plt.figure(figsize=(10, 6))
sns.barplot(x='Global_Sales', y='Genre', data=top_5_sales_by_genre, ci=None)
plt.xlabel('Global Sales')
plt.ylabel('Genre')
plt.title('Top 5 Global Sales by Genre')

plt.show()
```

```python
df_filtered = df[(df["Year_of_Release"] >= 2010) & (df["Year_of_Release"] <= 2014)]
```

```python
correlation = df_filtered["NA_Sales"].corr(df_filtered["JP_Sales"])
```

```python
print(f"Correlation between na_sales and jp_sales for the years 2010-2014: {correlation:.2f}")
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x="NA_Sales", y="JP_Sales", data=df_filtered)
plt.title("Scatter Plot: na_sales vs jp_sales (2010-2014)")
plt.xlabel("North America Sales")
plt.ylabel("Japan Sales")
plt.show()
```

```python
top_genres = df["Genre"].value_counts().nlargest(4)
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=top_genres.index, y=top_genres.values)
plt.xlabel("Genre")
plt.ylabel("Count")
plt.title("Top 4 Game Genres")
plt.show()
```
